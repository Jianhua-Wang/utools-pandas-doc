```python
import requests
from bs4 import BeautifulSoup
```

```python
main_url = 'https://pandas.pydata.org/pandas-docs/stable/reference'
response = requests.get(f'{main_url}/index.html')
soup = BeautifulSoup(response.content)
```

```python
for chapter in soup.select('.nav>li>a'):
    print(f'{chapter["href"]}\t{chapter.text}')
```

```python
chapter_url = f'{main_url}/{chapter["href"]}'
response = requests.get(chapter_url)
soup = BeautifulSoup(response.content)
```

```python
for method in soup.select('tr'):
    print(method)
```

```python
title,description = method.select('td')
```

```python
title_text = title.a['title']
title_url = title.a['href']
```

```python
response = requests.get(f'{main_url}/{title_url}')
soup = BeautifulSoup(response.content)
```
